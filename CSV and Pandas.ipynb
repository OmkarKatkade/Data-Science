```python
##Importing Pandas library as pd. 'pd' is an alias here. 
import pandas as pd

##Creating a panda dataframe and loading csv file's data into it. We are using an URL to load the csv. 
url = "https://support.spatialkey.com/wp-content/uploads/2021/02/SacramentocrimeJanuary2006.csv"
dframe = pd.read_csv(url)

##Made us of the Crime Records dataset. 
```

```python
## pandas' shape method return a tuple with the dimensions of the data set. Using this method to get the row and column count in the dataframe.
row_count, column_count = dframe.shape
print("The row count is: ",row_count)
print("The column count is: ",column_count)
```

```python
##Displaying first 8 rows of the dataframe using .head(n). n indicates the no.of rows we want to display.
dframe.head(8)
```

```python
##Displaying last 3 rows of the dataframe using .tail(n). n indicates the no.of rows we want to display.
dframe.tail(3)
```

```python
##Using pandas .T property to get the transpose of the dataframe. This .T property inturn calls .transpose() method to get the transpose. .T property return a dataframe which is transposed.
dframe_transpose = dframe.T

##Displaying first four rows' data (now column in our transposed dataframe)
dframe_transpose[dframe_transpose.columns[0:4]]
```
